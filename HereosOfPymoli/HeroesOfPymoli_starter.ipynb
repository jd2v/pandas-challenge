```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
player_count = len(purchase_data["SN"].unique())
player_count
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#find values
item_count = len(purchase_data["Item ID"].unique())
avg_cost = purchase_data["Price"].mean()
profits = purchase_data["Price"].sum()
purchase_count = purchase_data["Price"].count()

#make a summary df
summary_df = pd.DataFrame({"Unique Item Count": [item_count],
                            "Purchase Count": [purchase_count],
                            "Average Cost": [avg_cost],
                            "Profits": [profits]})
summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#find counts and ratio
player_genders = purchase_data.loc[:, ["Gender", "SN"]].drop_duplicates()
gender_totals = player_genders["Gender"].value_counts()
gender_percents =  (gender_totals / player_count)*100
gender_df = pd.DataFrame({"Players by Gender": gender_totals, "Percentage by Gender": gender_percents})
gender_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Run basic calculations
gender_purchases = purchase_data.groupby(["Gender"]).sum()["Price"].rename("Total Purchase Value")
gender_average = purchase_data.groupby(["Gender"]).mean()["Price"].rename("Average Purchase Price")
gender_counts = purchase_data.groupby(["Gender"]).count()["Price"].rename("Purchase Count")

# Calculate Normalized Purchasing (Average Total Purchase per Person)
avg_totals = gender_purchases / gender_totals

# make a df
gender_purchase_df = pd.DataFrame({"Purchase Count": gender_counts, "Average Purchase Price": gender_average, "Total Purchase Value": gender_purchases, "Average Player Spending per Purchase": avg_totals})


gender_purchase_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# make bins
age_bins = [0, 10, 15, 20, 25, 30, 35, 40, 99999]
bins_names = ["=<10", "11-15", "16-20", "21-25", "26-30", "31-35", "36-40", "41+"]

#put players into bins
player_ages = purchase_data.loc[:, ["Age", "SN"]].drop_duplicates()
player_ages["Age Groups"] = pd.cut(player_ages["Age"], age_bins, labels=bins_names)

# calculate totals and percents by age group
age_totals = player_ages["Age Groups"].value_counts()
age_percents = (age_totals / player_count)*100
age_df = pd.DataFrame({"Total Count": age_totals, "Percentage of Players": age_percents})

#reorder and show
age_df = age_df.sort_index()
age_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# put the ages into bins
purchase_data["Age Groups"] = pd.cut(purchase_data["Age"], age_bins, labels=bins_names)

# find summary stats
age_purchases = purchase_data.groupby(["Age Groups"]).sum()["Price"].rename("Total Purchases")
age_average = purchase_data.groupby(["Age Groups"]).mean()["Price"].rename("Average Purchase Cost")
age_counts = purchase_data.groupby(["Age Groups"]).count()["Price"].rename("Purchases Count")

# find average cost per perchase per age group
per_age_per_purchase = age_purchases / age_totals

#make the df
age_data = pd.DataFrame({"Purchase Counts": age_counts, "Average Purchase Price": age_average, "Total Purchase Price": age_purchases, "Avg Total per Player": per_age_per_purchase})


age_data
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#get summary stats
spender_totals = purchase_data.groupby(["SN"]).sum()["Price"].rename("Total Purcahse Price")
spender_avg = purchase_data.groupby(["SN"]).mean()["Price"].rename("Average Purchase Price")
spender_count = purchase_data.groupby(["SN"]).count()["Price"].rename("Purchase Count")

# make the spender df
spender_df = pd.DataFrame({"Total Money Spent": spender_totals, "Average Purchase Cost": spender_avg, "Number of Purcahses": spender_count})

#sort spender_df
sorted_spender_df =  spender_df.sort_values("Total Money Spent", ascending=False)

sorted_spender_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#grab items, names, and prices
item_df =  purchase_data.loc[:,["Item ID", "Item Name", "Price"]]

#group by items and find values
item_totals = item_df.groupby(["Item ID", "Item Name"]).sum()["Price"].rename("Total Money Spent on Item")
item_cost_avg = item_df.groupby(["Item ID", "Item Name"]).mean()["Price"]
item_counts = item_df.groupby(["Item ID", "Item Name"]).count()["Price"].rename("Number of Times Purchased")

# summary df and sort
item_summary_df = pd.DataFrame({"Total Money Spent on Item": item_totals, " Average Item Price": item_cost_avg, "Purchase Count": item_counts})
item_sorted_df = item_summary_df.sort_values("Purchase Count", ascending=False)

item_sorted_df.head()
```
